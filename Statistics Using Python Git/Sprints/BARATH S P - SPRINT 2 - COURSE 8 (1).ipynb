```python
def outliers(col):
    ser=pd.Series(df[col])
    q1=ser.quantile(0.25)
    q3=ser.quantile(0.75)
    iqr=q3-q1
    uf=q3+1.5*iqr
    lf=q1-1.5*iqr
    for z in ser:
        if z>uf or z<lf:
            df[col]=df[col].replace([z],df[col].mean())
```

```python
#TASK3
#10 15 30 200 1000 5000 10000 50000 100000
num = [10,15,30,200,1000,5000,10000,50000,100000]
data_s = []
fig,ax = plt.subplots(3,3,figsize=(14,10))
for i in num:
    sample_df=Fare.sample(n=i,replace=True,random_state=1)
    data_s.append(sample_df['fare_amount'])
k=0
for i in range(0,3):
    for j in range(0,3):
        ax[i,j].hist(data_s[k],bins=10,density=True)
        ax[i,j].set_title(label='Sample_size='+str(len(data_s[k])))
        k=k+1
plt.show()
```

As the sample size increases (e.g., 200, 1000, 5000, etc.), the distribution becomes smoother and starts to look more like a normal curve, thanks to the Central Limit Theorem.
By the time the sample size reaches 1000 or more, the distribution is very close to normal, even if the population doesn’t follow a perfect normal distribution.
Larger sample sizes reduce the effect of random fluctuations and provide a clearer and more accurate picture of the population's overall characteristics.

```python
# Mam Task 
sample_df=pd.DataFrame()
for i in range(0,20):
    sample_df["Sample_="+str(i)]=df["fare_amount"].sample(n=200,replace=True).reset_index(drop=True)
sample_df
```

```python
sample_name=sample_df.columns
ww=[]
for i in sample_name:
    sample_200_mean=st.mean(sample_df[i])
    sample_200_median=st.median(sample_df[i])
    sample_200_mode=st.mode(sample_df[i])
    sample_200_std=st.stdev(sample_df[i])
    sample_200_skew=skew(sample_df[i])
    sample_200_kurtosis=kurtosis(sample_df[i])
    ww.append({
    "columns":i,
    "Mean":sample_200_mean,
    "Median":sample_200_median,
    "Mode":sample_200_mode,
    "standard deviation":sample_200_std,
    "skew":sample_200_skew,
    "kurtosis":sample_200_kurtosis})
result=pd.DataFrame(ww)
result
```

INTERPRETATION
Mean & Median:

The mean and median of both the population and the sample (size 200) are nearly identical, indicating a balanced distribution without significant skewness.
Kurtosis & Skewness:

Kurtosis:
The sample has a slight bias in kurtosis, meaning it doesn't fully reflect the population's tail behavior.
The population's kurtosis (~0.6) suggests thinner tails, meaning fewer extreme values compared to a normal distribution.
Skewness:
Both the sample and the population are slightly positively skewed, meaning there’s a longer right-side tail in the data.
Standard Deviation of Sample Mean:

When taking 20 smaller samples, the variation (standard deviation) of their averages is close to the population’s standard deviation divided by the square root of the sample size. This aligns with statistical theory (Central Limit Theorem).
Distribution Shape:

The population and the sample of size 200 don’t have the classic bell-shaped curve (normal distribution) because of the thinner tails.
While the sample curve is not an exact match, it generally resembles the population curve, capturing the overall pattern.
